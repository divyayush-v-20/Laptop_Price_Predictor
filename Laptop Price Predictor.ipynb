```python
def get_os(text):
    if 'Windows' in text:
        return "Windows"
    elif 'Mac' or 'Apple' in text:
        return 'Mac Os'
    else:
        return 'Other/Linux/None'
```

```python
df['OpSys'] = df['OS'].apply(get_os)
```

#### RELATION B/W PROCESSOR BRAND AND PRICE

```python
def processor_brand(x):
    if 'Intel' in x:
        return 'Intel'
    elif 'AMD' in x or 'Ryzen' in x:
        return 'AMD'
    elif 'Apple' in x or 'Mac' in x or 'M1' in x or 'M2' in x:
        return 'Apple'
```

```python
df['processor_brand'] = df['processor'].apply(processor_brand)
```

```python
df.sample(5)
```

```python
sns.barplot(x = df['processor_brand'],y = df['price'])
plt.xticks(rotation = 90)
plt.show()
```

#### RELATION B/W GRAPHICS CARD BRAND AND PRICE

```python
def get_gpu_brand(gpu):
    if 'Intel' in gpu or 'Iris' in gpu:
        return 'Intel'
    elif 'AMD' in gpu or 'Radeon' in gpu:
        return 'AMD'
    elif 'GeForce' in gpu:
        return 'NVIDIA'
    elif 'core Graphics' in gpu:
        return 'Apple'
```

```python
df['gpu_brand'] = df['GPU'].apply(get_gpu_brand)
```

```python
df.sample(5)
```

```python
sns.barplot(x = df['gpu_brand'],y = df['price'])
plt.xticks(rotation = 90)
plt.show()
```

#### RELATION B/W SPEC RATING AND PRICE

```python
def get_spec_ranges(spec):
    if spec>=60 and spec<65:
        return '60-65'
    elif spec>=65 and spec<70:
        return '65-70'
    elif spec>=70 and spec<75:
        return '70-75'
    elif spec>=75 and spec<80:
        return '75-80'
    elif spec>=80:
        return 'above 80'
```

```python
df['spec_range'] = df['spec_rating'].apply(get_spec_ranges)
```

```python
df.sample(5)
```

```python
plt.figure(figsize = (10,7))
plt.scatter(df['spec_rating'],df['price'],alpha = 0.5)
plt.xlabel('Spec Rating of Laptop')
plt.ylabel('Price')
plt.show()
```
